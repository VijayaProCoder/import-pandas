### Introduction:
##### Data visualization is as much a part of the data processing step.. It is much easier to compare values when they are plotted than numeric values. By visualizing data we are able to get better understand of the data. Visualizations can find some hidden pattern and in-depth meaning in data.

#### There are 3 ways (libraries) we can generate visualizations:

#### 1. matplotlib.pyplot (more control on the graphs)
#### 2. seaborn (generates interactive visualizations). It also has some inbuilt datasets
#### 3. pandas to generate visualizations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("anscombe")
df
```

```python
type(df)
```

```python
df.columns
```

```python
dataset_1 = df[df['dataset'] == 'I']
```

```python
dataset_1
```

```python
plt.figure(figsize = (15,10))
plt.plot(dataset_1['x'], dataset_1['y'], 'r*')
```

```python
dataset_2 = df[df['dataset'] == 'II']
dataset_3 = df[df['dataset'] == 'III']
dataset_4 = df[df['dataset'] == 'IV']
```

```python
dataset_2
```

```python
dataset_3
```

```python
dataset_4
```

```python

fig = plt.figure(figsize = (10,8)) # creating an empty figure
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

# axes = a plot, axis = x-axis, y-axis\

axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'b*')
axes3.plot(dataset_3['x'], dataset_3['y'], 'r+')
axes4.plot(dataset_4['x'], dataset_4['y'], 'r.')


axes1.set_title("Dataset_1")
axes2.set_title("Dataset_2")
axes3.set_title("Dataset_3")
axes4.set_title("Dataset_4")

fig.suptitle("Anscombe Dataset plotting")
fig.tight_layout()
```
